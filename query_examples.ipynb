## _Indepedently generate query URLs._

```python
try_url = generate_url('Serena Williams', 'WTA',
                       {'surface': ['carpet', 'clay'], 'level': 'premier', 
                        'start date': datetime(1999, 8, 7),
                        'end date': datetime(2010, 4, 14),
                        'as rank': 'Top 5',
                        'round': ['R16', 'Round of 32']})
```

**in English**:
<br>
_Serena Williams matches in the Round of 32 or Round of 16 of Premier tournaments
on clay or carpet between 7-8-1999 and 4-14-2010 where she was ranked in the top 5._

```python
# query and download the data
try_stats = DownloadStats(try_url, tour='WTA')
```

```python
print(try_stats.title)
try_stats.stats.head(7)
```

```python
# generate a more convoluted query, this time for an ATP player
another_url = generate_url('Andy Murr', # don't need full name if there's only one match
                           'ATP',
                           {'vs height': 'Shorter', 'vs hand': 'right',
                            'vs entry': ['wild card', 'seeded'],
                            'as entry': 'seeded', 'vs current rank': 'inactive',
                            'vs rank': (14, 112),
                            'sets': ['straights', '4 of 5 sets'],
                            'score': 'all 7-6',
                            'exclude opp': ['Arnaud Clement', 'Guillermo Canas']})
```

**in English**:
<br>
_Andy Murray matches where he was seeded -- and his opponent (who is not Arnaud Clement 
or Guillermo Canas) was right-handed, shorter than him, ranked between 14-112 at the time,
was either seeded or given a wild card, and is currently inactive -- that ended in
4 sets or straight sets and featured at least 1 tiebreak._

```python
another_stats = DownloadStats(another_url, tour='ATP')
```

```python
print(another_stats.title)
another_stats.stats
```
